```python
import ginsim
import boolsim
import bns
import os
from colomoto import minibn
import biolqm
from colomoto_jupyter import tabulate
import csv
import pandas as pd
import numpy as np
import time
```

```python
current_path=os.getcwd()
```

## The Cardiomyocyte BN

```python
model = minibn.BooleanNetwork.load("CardiomyocyteBN.bnet")
model
lqm2 = biolqm.load("CardiomyocyteBN.bnet")
lrg = biolqm.to_ginsim(lqm2)
ginsim.show(lrg)
# Asynchronous Attractors
a = boolsim.attractors(model.to_biolqm())
tabulate(a)
```

```python
ginsim.show(lrg)
```
